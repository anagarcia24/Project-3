```python
import pandas as pd
from pathlib import Path

data = Path("Resources/NPSAcreage-09-30-2024.xlsx")
df = pd.read_excel(data) 
df
```

```python
# Delete the first row
df = df.iloc[0:]

# Set the new header
df.columns = df.iloc[0]

# Drop the row that was used as the header
df = df.iloc[1:]

# Reset the index
df.reset_index(drop=True, inplace=True)

df
```

```python
#drop rows that have all Null information
clean_df = df.dropna(subset=df.columns[1:], how='all')
clean_df
```

```python
#Sort the dataframe from largest to smallest gross area acres

df_sorted = clean_df.sort_values(by='Gross Area Acres', ascending=False)
df_sorted
```
