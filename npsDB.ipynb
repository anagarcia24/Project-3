```python
import pandas as pd
from pathlib import Path

data = Path("Resources/allParksVisits_try.csv")
df = pd.read_csv(data)
df
```

```python
#Convert all datatypes from strings to integers
df['RecreationVisitors'] = df['RecreationVisitors'].str.replace(',', '').astype(int)
df['NonRecreationVisitors'] = df['NonRecreationVisitors'].str.replace(',', '').astype(int)
df['RecreationHours'] = df['RecreationHours'].str.replace(',', '').astype(int)
df['NonRecreationHours'] = df['NonRecreationHours'].str.replace(',', '').astype(int)
df['ConcessionerLodging'] = df['ConcessionerLodging'].str.replace(',', '').astype(int)
df['ConcessionerCamping'] = df['ConcessionerCamping'].str.replace(',', '').astype(int)
df['TentCampers'] = df['TentCampers'].str.replace(',', '').astype(int)
df['RVCampers'] = df['RVCampers'].str.replace(',', '').astype(int)
df['Backcountry'] = df['Backcountry'].str.replace(',', '').astype(int)
df['NonRecreationOvernightStays'] = df['NonRecreationOvernightStays'].str.replace(',', '').astype(int)
df['MiscellaneousOvernightStays'] = df['MiscellaneousOvernightStays'].str.replace(',', '').astype(int)

df.info()
```
