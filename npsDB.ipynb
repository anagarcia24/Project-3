```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

# Define the path to the CSV file
data = Path("Resources/allParksVisits_try.csv")

# Check if the file exists
if data.exists():
    # Read the CSV file into a DataFrame
    df = pd.read_csv(data)
    # Display the first few rows of the DataFrame
    print(df.head())
else:
    print(f"File not found at {data}")
```

```python
df.head(200)
```

```python
#Convert all datatypes from strings to integers
df['RecreationVisitors'] = df['RecreationVisitors'].str.replace(',', '').astype(int)
df['NonRecreationVisitors'] = df['NonRecreationVisitors'].str.replace(',', '').astype(int)
df['RecreationHours'] = df['RecreationHours'].str.replace(',', '').astype(int)
df['NonRecreationHours'] = df['NonRecreationHours'].str.replace(',', '').astype(int)
df['ConcessionerLodging'] = df['ConcessionerLodging'].str.replace(',', '').astype(int)
df['ConcessionerCamping'] = df['ConcessionerCamping'].str.replace(',', '').astype(int)
df['TentCampers'] = df['TentCampers'].str.replace(',', '').astype(int)
df['RVCampers'] = df['RVCampers'].str.replace(',', '').astype(int)
df['Backcountry'] = df['Backcountry'].str.replace(',', '').astype(int)
df['NonRecreationOvernightStays'] = df['NonRecreationOvernightStays'].str.replace(',', '').astype(int)
df['MiscellaneousOvernightStays'] = df['MiscellaneousOvernightStays'].str.replace(',', '').astype(int)

df.info()
```

```python
parkNames = df['ParkName'].unique()
```

```python
df['ParkName'] = df['ParkName'].astype(str)
df['ParkName'] = df['ParkName'].str.replace('NP', 'National Park')
df['ParkName'] = df['ParkName'].str.replace('NHP', 'National Historical Park')
df['ParkName'] = df['ParkName'].str.replace('MEM', 'Memorial')
df['ParkName'] = df['ParkName'].str.replace('NMEM', 'National Memorial')
```

```python
parkNames 
```

```python
# Set the year filter (for a range of years: 2010 to 2020)
start_year = 2010
end_year = 2020
filtered_df = df[(df['Year'] >= start_year) & (df['Year'] <= end_year)]

# Group by park and sum the visitors
park_visitors = filtered_df.groupby('ParkName')['RecreationVisitors'].sum().sort_values(ascending=False)

# Plot the top 10 parks by recreation visitors
plt.figure(figsize=(12, 8))
park_visitors.head(10).plot(kind='bar', color='skyblue')
plt.title(f'Top 10 Parks by Recreation Visitors ({start_year} to {end_year})')
plt.xlabel('Park Name')
plt.ylabel('Total Recreation Visitors')
plt.xticks(rotation=90)
plt.show()


```

```python
# Filter the data for a specific park (e.g., "Abraham Lincoln Birthplace NHP")
park_name = 'Abraham Lincoln Birthplace National Historical Park'
park_data = df[df['ParkName'] == park_name]

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(park_data['Year'], park_data['RecreationVisitors'], marker='o', color='green')
plt.title(f'Recreation Visitors Over Time: {park_name}')
plt.xlabel('Year')
plt.ylabel('Recreation Visitors')
plt.grid(True)
plt.show()
```

```python
# Grouping by park and summing both types of visitors
park_visitors_comparison = df.groupby('ParkName')[['RecreationVisitors', 'NonRecreationVisitors']].sum()

# Plotting the stacked bar chart
park_visitors_comparison.head(10).plot(kind='bar', stacked=True, figsize=(12, 8), color=['lightgreen', 'lightcoral'])
plt.title('Top 10 Parks: Recreation vs Non-Recreation Visitors')
plt.xlabel('Park Name')
plt.ylabel('Visitors')
plt.xticks(rotation=90)
plt.show()
```

```python
#A scatter plot to visualize the relationship between RecreationHours and RecreationVisitors for each park.
plt.figure(figsize=(10, 6))
plt.scatter(df['RecreationHours'], df['RecreationVisitors'], alpha=0.6, color='purple')
plt.title('Recreation Hours vs. Recreation Visitors')
plt.xlabel('Recreation Hours')
plt.ylabel('Recreation Visitors')
plt.grid(True)
plt.show()
```

```python
# Choose a specific park
park_name = 'Zion National Park'
park_data = df[df['ParkName'] == park_name]

# Summing visitor types for the park
visitor_types = park_data[['RecreationVisitors', 'NonRecreationVisitors']].sum()

# Plotting the pie chart
plt.figure(figsize=(7, 7))
visitor_types.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title(f'Visitor Proportions: {park_name}')
plt.ylabel('')
plt.show()
```

```python
#6. Histogram: Distribution of Recreation Visitors
plt.figure(figsize=(10, 6))
plt.hist(df['RecreationVisitors'], bins=30, color='orange', edgecolor='black')
plt.title('Distribution of Recreation Visitors')
plt.xlabel('Recreation Visitors')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#7. Bar Chart: Concessioner Lodging by Year
# Choose a specific park
park_name = 'Zion National Park'
park_data = df[df['ParkName'] == park_name]

# Plotting the bar chart for Concessioner Lodging by Year
plt.figure(figsize=(10, 6))
plt.bar(park_data['Year'], park_data['ConcessionerLodging'], color='teal')
plt.title(f'Concessioner Lodging Over the Years: {park_name}')
plt.xlabel('Year')
plt.ylabel('Concessioner Lodging')
plt.grid(True)
plt.show()
```

```python
#8. Heatmap: Correlation Between Numeric Variables
import seaborn as sns

# Select numeric columns only
numeric_columns = df.select_dtypes(include=['int64', 'float64'])

# Compute the correlation matrix
correlation_matrix = numeric_columns.corr()

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Numeric Columns')
plt.show()
```
